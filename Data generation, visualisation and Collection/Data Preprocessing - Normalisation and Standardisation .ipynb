```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# the following is a linear regression data
dfx = pd.read_csv("./linearX.csv")
dfy = pd.read_csv("./linearY.csv")
print(dfx.shape, dfy.shape)
print(type(dfx))

# converting dataframe type to numpy array in order to operate on it
x = dfx.values
y = dfy.values
print(x.shape, y.shape)
print(type(x))

plt.scatter(x,y,label = "Raw data")

# Since the spread is more so Normalising the data
x2 = (x-x.mean())/x.std()
plt.scatter(x2, y, color = "orange", label = "Normalised Data")
plt.legend(loc = "upper left", bbox_to_anchor = (1,1))
plt.show()
# this brings the mean of the data to 0 and std deviation as 1
# the new dataset has mean = 0 and deviation(spread) from the mean as 1
```
