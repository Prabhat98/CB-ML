```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,3,4,5])
y = np.array([4,5,6,7,8])
a = x**2
b = x*10
y = y*5
plt.plot(a,y, label = "Curve")
plt.plot(b,y, label = "Line")
plt.legend(loc = "upper left", bbox_to_anchor = (1,1))
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("The graph")
# plt.show() gets rid of the statement written before the graph after doing plt.plot()
plt.show()
```

## Bar Graph

```python
numbers = np.random.randint(1, 10, 5)
# range() gives the list of numbers from 0 to n-1
indices = np.arange(5)
print(numbers)
print(indices)
# in case of bar the x-axis will become the indices and y-axis will become the numbers as supplied
# the numbers become the height of the bar
# the numerical value tells the width of the bar
plt.bar(indices, numbers, 0.5, color = "red", label = "Graph 1")
plt.legend()
plt.show()
```

## Pie Chart

```python
# lets say for example we need to plot a pie chart that represents the prerequisites for machine learning
labels = ["Algebra", "Calculus", "Coding", "Algorithms"]
# sum can be 100 of values but not required, not necessary, when the sum is not 100 then it automatically calculates the
# percentage of each label
values = [90, 80, 54, 40]
# in order to highlight a particular section we will assign that particular sector as 1 and the remaing as 0 in the tuple
plt.pie(values, labels = labels, radius = 1, explode = (1, 0, 0, 0), shadow = True)
plt.show()
```
