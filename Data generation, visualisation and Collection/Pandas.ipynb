## Pandas 
### Pandas is an open source library used for data analysis in python, used for reading and writing data

```python
import pandas as pd
# we will have to read the file. since it is a csv file so read_csv else there are many methods to read different types of files

df = pd.read_csv("./movie_metadata.csv")
df.head(10) # by default shows only fist five rows, putting a value gives those many samples having 28 features
df
```

```python
print(df.columns) # gives us columns
# to get a particular column
df['imdb_score'] # or
score = df.get('imdb_score') # both mean the same
type(score) # not exactly a python list but can be converted into one
score = list(score)
type(score)
```

```python
type(df) # it is a type of dataframe
# to convert the csv file into numpy array to apply operations on it
data = df.values # values attribute gives numpy array of the data
print(type(data))
print(data.shape) # 5043 rows and 28 columns
```

```python
import matplotlib.pyplot as plt
# Visualisation based on some insights of the data
# X axis will have length of the movie name, Y-axis will have the frequency of the length of the name
titles = df['movie_title']
type(titles)

# creating an empty dictionary
dic = {}

# we will iterate over all the titles and compute their length and maintain a dictionary

# if else coouldn't be used cos condition can't be derived to be put in if 
for t in titles:
    length = len(t)
    # if key already present then try block will run
    try:
        # incrementing already present keys
        dic[length] = dic[length] + 1
    # if key not present then except block will run
    except:
        # except handles exceptions raised by try block
        # inserting new keys
        dic[length] = 1
print(dic.keys())
print(dic.values())

#keys go on x-axis and values go on y-axis
plt.scatter(dic.keys(), dic.values())
plt.xlabel("Length of movie title")
plt.ylabel("Frequency of lengths")
plt.show()    

# the plot can be related to some kind of normal distribution with some mean and sigma, hence most distribution in the 
# real life are normal
```

## MNIST dataset visualisation
### it is a dataset of hand written images

```python
import pandas as pd
df = pd.read_csv("./mnist_train.csv")
```

```python
# this data set contains pixel values just like load_digits from sklearn library, the only difference is that instead of
# 8X8 matrix to represent pixel values of an image(of a number) there is a 28X28 matrix that has been formed into a linear array
# and there are 42000 examples of such images, the first column represents the value that image represents(the number itself)
print(df.shape)
# here the x values and the y values are combined together so wew need to glean them out and separate them 
# the x value is the examples with features (actual data) and y values are the target (the value that image represents)
```

```python
# the first calumn is the label and the rest 784 columns are for pixels
print(type(df))
# converting pandas dataframe into numpy array in order to perform operations
data = df.values
print(type(data))

# sliced n dimensional array and stored data values in x and target values in y 
x = data[:,1:]
y = data[:,0]
print(x.shape)
print(y.shape)

# tool 899th example and reshaped in 28X28 matrix
matrix = x[899].reshape(28,28)

import matplotlib.pyplot as plt
plt.imshow(matrix, cmap = "gray")
plt.show()

print(y[899])
```

```python
# Now dividing the given dataset into training set and test set
# will given 80% as training data and 20% as testing data

# x.shape is a tuple with two elements so we are accessing the first element(42000) and getting its 80%
split = int(.80*x.shape[0]) # rounding off to nearest integer

# now assigning the values, in xtrain we need all the columns and rows(samples) upto 80% of total i.e split 
xtrain, ytrain = x[:split, :], y[:split]
print(xtrain.shape)
print(ytrain.shape)

# the remaing examples of the dataset will go to test set
xtest, ytest = x[split:, :], y[split:]
print(xtest.shape)
print(ytest.shape)

# we've divided our data set into two parts, one for training purpose and the other one for testing purpose
```

## Or we can use sklearn inbuilt function for splitting the data
