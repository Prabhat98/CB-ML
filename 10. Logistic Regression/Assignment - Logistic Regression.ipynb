```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Preparing data

```python
x_data = pd.read_csv("./data/logisticX.csv")
y_data = pd.read_csv("./data/logisticY.csv")
x_train = x_data.values
y_train = y_data.values

print(x_train.shape, y_train.shape)
```

## Visualising the data

```python
plt.scatter(x_train[:,0], x_train[:, 1])
plt.show()
```

## Algorithm

```python
def hypothesis(theta, x_train):
    y = theta[0] + theta[1]*x_train[0] + theta[2]*x_train[1]
    sigmoid = 1.0/(1.0 + np.exp(-y))
    return sigmoid

def mini_batch_gradient(x_train, y_train, theta, batch_size = 10):
    m = x_train.shape[0]    
    ind = np.arange(m)
    np.random.shuffle(ind)
    indices = ind[:batch_size]
    grad = np.zeros((3, ))
    for i in indices:
        grad[0] = grad[0] + (y_train[i] - hypothesis(theta, x_train[i]))
        grad[1] = grad[1] + (y_train[i] - hypothesis(theta, x_train[i]))*x_train[i][0]
        grad[2] = grad[2] + (y_train[i] - hypothesis(theta, x_train[i]))*x_train[i][1]
    return grad

def mini_batch_update(x_train, y_train, learning_rate, max_iterations):
    theta = np.zeros((3, ))
    combined_error = []
    for i in range(max_iterations):
        grad = mini_batch_gradient(x_train, y_train, theta)
        err = error(x_train, y_train, theta)
        theta[0] = theta[0] + learning_rate*grad[0]
        theta[1] = theta[1] + learning_rate*grad[1]
        theta[2] = theta[2] + learning_rate*grad[2]
        combined_error.append(err)
    return theta, combined_error

def error(x_train, y_train, theta):
    m = x_train.shape[0]
    error = 0.0
    for i in range(m):
        hx = hypothesis(theta, x_train[i])
        error = error + (y_train[i])*(np.log(hx)) + (1 - y_train[i])*(np.log(1 - hx))
    return -error
```

```python
theta, combined_error = mini_batch_update(x_train, y_train, learning_rate = 0.01, max_iterations = 800)
print(theta, min(combined_error))
plt.plot(combined_error)
plt.show()
```

```python
def predicted_y(theta, x_train):
    confidence = np.zeros((x_train.shape[0]))
    m = x_train.shape[0]
    for i in range(m):
        con = hypothesis(theta, x_train[i])
        if(con < 0.5):
            confidence[i] = 0
        else:
            confidence[i] = 1
    return confidence

def accuracy(x_train, y_train, theta):
    m = y_train.shape[0]
    accuracy = []
    y_predicted = predicted_y(theta, x_train)
    count = 0.0
    for i in range(m):
        if(y_predicted[i] == y_train[i]):
            count = count + 1
            acc = count/m
            accuracy.append(acc)
    return accuracy

acc = accuracy(x_train, y_train, theta)
print("Final accuracy of model achieved by mini batch gradient descent is", acc[-1])

plt.plot(acc)
plt.ylabel("Accuracy")
plt.xlabel("Number of iterations")
plt.show()
```

```python
def boundary(x_train, theta):
    m = x_train.shape[0]
    x2 = np.zeros((m, ))
    for i in range(m):
        x2[i] = -((theta[0] + theta[1]*x_train[i][0])/theta[2])
    return x2

x2 = boundary(x_train, theta)

plt.scatter(x_train[:,0], x_train[:, 1], label = "Training data")
plt.plot(x_train[:, 0], x2, color = "red", label = "Boundary")
plt.legend(loc = "upper left", bbox_to_anchor = (1, 1))
plt.show()
```

## Using sklearn to compare models
