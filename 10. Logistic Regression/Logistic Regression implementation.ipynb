```python
import numpy as np
import matplotlib.pyplot as plt
```

## Preparing the data

```python
# defined mean and covariance for the first dataset
mean_1 = np.array([1, 0.5])
cov_1 = np.array([[1, 0.1],[0.1, 1.2]])

# defined mean and covariance for the second dataset
mean_2 = np.array([4, 5])
cov_2 = np.array([[1.21, 0.1],[0.1, 1.3]])

# multivariate_normal draws out samples from the mulivariate distribution(that has more that one feature)
dist_1 = np.random.multivariate_normal(mean_1, cov_1, 500)
dist_2 = np.random.multivariate_normal(mean_2, cov_2, 500)

# distribution 1 and 2 will have 500 examples each with two features x1 and x2
print(dist_1.shape)
print(dist_2.shape)
```

## Data visualisation

```python
# x1 feature goes on the x axis and x2 feature goes on y axis
plt.scatter(dist_1[:, 0], dist_1[:, 1], label = "Class 0")
plt.scatter(dist_2[:, 0], dist_2[:, 1], label = "Class 1", marker = "^")
plt.legend(loc = "upper left", bbox_to_anchor = (1,1))
plt.show()

# the goal of our algorithm will be to generate a boundary the would separate values of class 0 from those of class 1 
# by using the concept of maximum likelihood(by generating opitmum value of paramters)
```

## Preparing training and testing dataset

```python
# data array will be our entire dataset, 1st column for x1 feature and 2nd column for x2 feature and 3rd column will have the 
# target values(i.e either 0 or 1)
data = np.zeros((1000, 3))
#print(data.shape)

# first 500 examples will be distribution 1 and remaining 500 examples will be distribution 2, half of them will have target 
# values as 0 and remaining half will have 1
data[:500, :2] = dist_1
data[500:, :2] = dist_2
data[500:, -1] = 1.0

np.random.shuffle(data)
#print(data[:10, :])

split = int(0.8*data.shape[0])

x_train = data[:split, :2]
y_train = data[:split, 2]
print(x_train.shape, y_train.shape)

x_test = data[split:, :2]
y_test = data[split:, 2]
print(x_test.shape, y_test.shape)
```

# The alogrithm

```python
def hypothesis(theta, x_train):
    y = theta[0] + theta[1]*x_train[0] + theta[2]*x_train[1]
    sigmoid_y = 1.0/(1.0 + np.exp(-1.0*y))
    return sigmoid_y

def error(x_train, y_train, theta):
    # log of likelihood function[log(l(theta))] will be a concave function which we need to maximize, negative of that 
    # function will be our loss/error function which needs to be minimized
    m = x_train.shape[0]
    total_error = 0
    for i in range(m):
        hx = hypothesis(theta, x_train[i])
        total_error = total_error + (y_train[i]*np.log(hx)) + (1 - y_train[i])*(np.log(1 - hx))
    # either we can take the total error or we can take the average error over each example, we are taking the average error
    return -(total_error)/m

def gradient(x_train, y_train, theta):
    grad = np.zeros((3,))
    m = x_train.shape[0]
    for i in range(m):
        grad[0] = grad[0] + (y_train[i] - hypothesis(theta, x_train[i]))
        grad[1] = grad[1] + (y_train[i] - hypothesis(theta, x_train[i]))*x_train[i][0]
        grad[2] = grad[2] + (y_train[i] - hypothesis(theta, x_train[i]))*x_train[i][1]
    return grad

def update_rule(x_train, y_train, learning_rate, max_iterations):
    combined_error = []
    m = x_train.shape[0]
    #theta = np.zeros((3,))
    theta = 2*np.random.random((3, ))
    for i in range(max_iterations):
        grad = gradient(x_train, y_train, theta)
        err = error(x_train, y_train, theta)
        theta[0] = theta[0] + learning_rate*grad[0]
        theta[1] = theta[1] + learning_rate*grad[1]
        theta[2] = theta[2] + learning_rate*grad[2]        
        combined_error.append(err)
    return theta, combined_error

theta, combined_error = update_rule(x_train, y_train, learning_rate = 0.001, max_iterations = 1000)
print(theta)

print(min(combined_error))

# Since we considered loss as -ve of log likelihood so the loss is decreasing, and as we increase max_iterations count,
# the loss tends to more and more less value in the end
plt.plot(combined_error)
plt.xlabel("Iteration count, Time")
plt.ylabel("Loss (Negative of log likelihood)")
plt.show()
    
```

```python
# Rounding the predicted y values to 0 or 1, because ultimately we want to separate the data into either class 0 or 1
def predicted_y_values(theta, x_train):
    confidence = np.zeros((x_train.shape[0], ))
    con = 0.0
    for i in range(x_train.shape[0]):
        con = hypothesis(theta, x_train[i])
        if(con < 0.5):
            confidence[i] = 0
        else:
            confidence[i] = 1
    return confidence

confidence = predicted_y_values(theta, x_train)
print(confidence.shape)

# In accuracy we are comparing the our predicted values to the actual values and computing the percentage of correct values
def accuracy(confidence, y_train):
    accu = []
    m = y_train.shape[0]
    sum = 0.0
    for i in range(m):
        if(confidence[i] == y_train[i]):
            sum = sum + 1
            acc = sum/m
            accu.append(acc)
    return accu

accuracy = accuracy(confidence, y_train)
print(type(accuracy))

# Calculated accuracy after every iteration and plotted the graph
plt.plot(accuracy)
plt.show()

print("Final accuracy of training model", accuracy[-1])
    
```

## Visualising Decision Boundary

```python
# Since our hypothesis is theta_0 + theta_1*x_1 + theta_2*x_2 and this would give us a y value and we want to plot a graph
# that would represend the given data (that is between x1 feature and x2 feature) and our boundary so we manupulated our 
# hypothesis equation for x2 to be on one side(which we'd calculate intead of y) and rest variables on the other side

def boundary(x_train, theta):
    m = x_train.shape[0]
    x2 = np.zeros((x_train.shape[0], ))
    for i in range(m):
        x2[i] = -((theta[0] + theta[1]*x_train[i][0])/theta[2])
    return x2

# So we get the decision boundary that our algorithm has generated and accuracy is almost 99%
x2 = boundary(x_train, theta)

plt.scatter(dist_1[:, 0], dist_1[:, 1], label = "Class 0")
plt.scatter(dist_2[:, 0], dist_2[:, 1], label = "Class 1", marker = "^")
plt.plot(x_train[:, 0], x2, label = "Decision Boundary", color = "green")
plt.legend(loc = "upper left", bbox_to_anchor = (1,1))
plt.show()    
```

## Calculating r2_score and theta using sklearn and comparing with our model

```python
# Computing the accuracy through r2score method
from sklearn.metrics import r2_score
y_predicted = np.zeros((800, ))
m = y_predicted.shape[0]
for i in range(m):
    y_predicted[i] = hypothesis(theta, x_train[i])
print("r2 score", r2_score(y_train, y_predicted))

y_t_predicted = np.zeros((200, ))
m = y_t_predicted.shape[0]
for i in range(m):
    y_t_predicted[i] = hypothesis(theta, x_test[i])
print("r2 score on test", r2_score(y_test, y_t_predicted))
```

```python
# Fitting the data using LogisticRegression function and checking how much our model matched with that of sklearn's
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)
print("intercept", lr.intercept_)
print("coeff", lr.coef_)

theta1 = np.zeros((3, ))
theta1[0] = lr.intercept_
theta1[1] = lr.coef_[0][0]
theta1[2] = lr.coef_[0][1]
print(theta1)
```

```python
from sklearn.metrics import r2_score
y2_predicted = np.zeros((800, ))
m = y2_predicted.shape[0]
for i in range(m):
    y2_predicted[i] = hypothesis(theta1, x_train[i])
print("r2 score", r2_score(y_train, y2_predicted))

y_t2_predicted = np.zeros((200, ))
m = y_t2_predicted.shape[0]
for i in range(m):
    y_t2_predicted[i] = hypothesis(theta1, x_test[i])
print("r2 score on test", r2_score(y_test, y_t2_predicted))
```
