```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,3,4,5])
y = np.array([4,5,6,7,8])
a = x**2
b = x*10
y = y*5
plt.plot(a,y, label = "Curve")
plt.plot(b,y, label = "Line")
plt.legend(loc = "upper left", bbox_to_anchor = (1,1))
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("The graph")
# plt.show() gets rid of the statement written before the graph after doing plt.plot()
plt.show()
```

## Bar Graph

```python
numbers = np.random.randint(1, 10, 5)
# range() gives the list of numbers from 0 to n-1
indices = np.arange(5)
print(numbers)
print(indices)
# in case of bar the x-axis will become the indices and y-axis will become the numbers as supplied
# the numbers become the height of the bar
# the numerical value tells the width of the bar
plt.bar(indices, numbers, 0.5, color = "red", label = "Graph 1")
plt.legend()
plt.show()
```

## Pie Chart

```python
# lets say for example we need to plot a pie chart that represents the prerequisites for machine learning
labels = ["Algebra", "Calculus", "Coding", "Algorithms"]
# sum can be 100 of values but not required, not necessary, when the sum is not 100 then it automatically calculates the
# percentage of each label
values = [90, 80, 54, 40]
# in order to highlight a particular section we will assign that particular sector as 1 and the remaing as 0 in the tuple
plt.pie(values, labels = labels, radius = 1, explode = (1, 0, 0, 0), shadow = True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# the value in randn specifies the number of points we want from the normal distribution
val = np.random.randn(100)
# in order to specify our mean and standard deviation we will
mean = 5
std = 2
# i.e shifted the whole curve about mean and multiplied the whole thing by 2
vals = mean + std*np.random.randn(100)
print(vals.shape)
#print(vals)
# the second value tells the number of bars that we want
# on x axis we have the points obtained from randn and on y axis we have the number of occurences of that point
plt.hist(vals, 50)
plt.show()
values = np.round(vals)
z = np.unique(values, return_counts = True)
print(z)
```

```python
x = vals
# we have plotted the points around a single line i.e y=0, the frequency can be seen by the density of the points in a region
# zeroes takes order of the matrix so will be same as shape of x, we took y = 0 line so np.zeros
y = np.zeros(x.shape)
plt.scatter(x,y)
plt.show()
```

```python
# a standard normal distribution has mean = 0 and standard deviation as 1, randn gives us numbers from the same normal 
# distribution, such a distribution is also called standard normal distribtion
import numpy as np
from numpy.random import randn
count = 0
point = 0
# As we increase the range our percentage reaches close to 68.2% as expected in normal distribution for -1 to +1
for i in range(100000):
    x = randn()
    count = count + 1
    if x <= 4 and x >= -4:
        point = point + 1 
print("The number of", count, "samples in -1 to +1 is", point, "and the count is", count)
precentage = (point/count)
print("The required percentage is", precentage*100, "for", count, "samples")
# around 68% of the total population lies within 1st standard deviation, around 95% of the population falls within the 
# two standard deviation and so on 
```

## Multivariate Normal Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
# bivariate
# considered that distribution is centered around origin
mean = np.array([0.0, 0.0])
# considered covariance as following
# standard deviation along x and y axis taken as 1
cov = np.array([[1,0],[0,1]])

mean2 = np.array([7.0,8.0])
cov2 = np.array([[1.3,-0.8],[-0.8,2.1]])
vals2 = np.random.multivariate_normal(mean2, cov2, 1000)

# random.multivariate_normal() draws samples from multivariate normal distribution, it takes mean, cov and the number of samples
# to be generated

vals = np.random.multivariate_normal(mean, cov, 1000)
# has 500 samples with two features and we are going to plot graph taking those two features on x and y axis
print(vals.shape)
# [:,0] means for all rows of 0th column on x axis, [:,1] for all rows of 1st column on y axis
plt.scatter(vals[:,0],vals[:,1])
plt.scatter(vals2[:,0],vals2[:,1])
plt.show()

```
