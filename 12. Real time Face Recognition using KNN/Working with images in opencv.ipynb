```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Reading the image using cv2
img = cv2.imread("./qq.jpg")
plt.imshow(img)
plt.show()
```

```python
# Since cv2 uses BGR by default so blue appears in images in place of red. Therefore converting BGR color to RGB
newimg = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(newimg)
plt.show()

# img.shape is a 3d matrix with 1080 rows, 1920 columns and 3 depth (rgb channels)
print(img.shape)
```

```python
# Shwoing image by just using opencv and not matplotlib

# Using cv2 we need not convert into color spaces 
cv2.imshow("Random Image", img) # This contains title of the window in which the image will be displayed and the image itself
# waitkey(0) means wait for infinite time untill it gets destroyed, waitkey(25) would mean that window will go off
# in 25 miliseconds
# waitkey basically means wating for specified period of time before next iteration comes up
cv2.waitKey(0)
# after the time in waitkey gets depleted then destroy all the windows
cv2.destroyAllWindows()
```

```python
# To read the image as grayscale image
gray = cv2.imread("./qq.jpg", cv2.IMREAD_GRAYSCALE) # cv2.IMREAD_GRAYSCALE is a constant that tells image to be read as grayscale
cv2.imshow("Gray", gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
