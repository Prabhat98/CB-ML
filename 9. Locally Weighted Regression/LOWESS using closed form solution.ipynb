```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = pd.read_csv("./data/weightedX.csv")
y = pd.read_csv("./data/weightedY.csv")

print(x.shape, y.shape)

# Normalising the data

x = (x - x.mean())/x.std()
```

```python
plt.scatter(x, y)
plt.show()
```

```python
# the only difference between a matrix and an array is that 
# a, b are matrices then a*b will give dot product
# a, b are arrays then a*b will give element wise product, dot product can be achieved by np.dot(a, b)

def weights(x, x_query, tau):
    m = x.shape[0] 
    # w can be conveted to matrix by w = np.mat(w) for multiplication pusposes or can be continued to be use as array
    # in that case we will have to use the dot method for multiplication 
    w = np.eye(m)
    
    for i in range(m):
        numerator = np.dot((x[i] - x_query), (x[i] - x_query).transpose())
        denomenator = 2*(tau**2)
        w[i,i] = np.exp((-numerator)/denomenator)
    return w

def prediction(x, y, x_query, tau):
    ones = np.ones((x.shape[0], 1))
    # hstack stacks arrays column wise
    # introduced one more feature to out data set x_not which will always be one, so that out hypothesis will become
    # Q[0]x[0] + Q[1]x[1] where x[0] will be 1
    x_new = np.mat(np.hstack((x, ones)))
    # simillarly if there are n features then our query point will alwso have (n+1) features where (n+1)th feature will be 1
    x_query_new = np.mat([x_query, 1])    
    w = weights(x_new, x_query_new, tau)
    y = np.mat(y)
    
    theta = (np.linalg.inv((x_new.transpose())*(w*x_new)))*((x_new.transpose())*(w*y))
    # we get a vector of two thetas since there is one feature so two parameters learned, theta_not and theta_one    
    return theta

def hypothesis(x_query, theta):
    # or x_query_new = np.array([x_query, 1]) will also work fine coz we have used dot method to perform multiplication
    x_query_new = np.mat([x_query, 1])
    y = np.dot(x_query_new, theta)
    return y

def plotplrediction(x, y, tau):
    # linspace will generate 10 points between -1.75 and 1.5 that are equally spaced
    # x_test will be our testing points to check how much well our prediction model works
    x_test_query = np.linspace(-1.75, 1.5, 20)
    y_predicted = []
    
    for i in x_test_query:
        theta = prediction(x, y, i, tau)
        new_y = hypothesis(i, theta)
        y_predicted.append(new_y)
        
    y_predicted = np.array(y_predicted) 
    
    plt.title("Tau/Bandwidth Parameter %.2f"%tau)
    plt.scatter(x, y, color = "yellow", label = "Actual Data")
    plt.scatter(x_test_query, y_predicted, color = "red", label = "Predicted Data")
    plt.legend(loc = "upper left", bbox_to_anchor = (1,1))
    plt.show()

plotplrediction(x, y, 0.1)
```

```python
# Plotting for different values of tau
tau = [0.1, 0.5, 1, 5, 10, 50, 100]

# as tau increases our algorithm moves more towards linear regression, as we give more bandwith we are giving more weightage
# to more number of points in finding out the paramters of the query point

# tau needs to be less as we want only neighbouring points to have more weight(that would affect the parameters of query point)
# and less weight to far off points 

for i in tau:
    plotplrediction(x, y, i)
```
