```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression

# Generated linear regression dataset

x, y = make_regression(n_samples = 400, n_features = 1, n_informative = 1, noise = 1.8, random_state = 11)
y = y.reshape((400, 1))
print(x.shape, y.shape)

# Normalising the given dataset

x = (x - x.mean())/x.std()
plt.scatter(x, y)
plt.show()
```

```python
# Creating a column of ones in our x matrix since we need a x_not feature to simplify our calculations

ones = np.ones((x.shape[0], 1))
x_new = np.mat(np.hstack((x, ones)))
print(x_new.shape)

def predict(x, theta):
    y_predicted = np.dot(x, theta)
    return y_predicted

def thetaclosedform(x, y):
    firstpart = np.dot(x.transpose(), x)
    secondpart = np.dot(x.transpose(), y)
    theta = (np.linalg.inv(firstpart))*(secondpart)
    return theta

theta = thetaclosedform(x_new, y)
print(theta)

y_predicted = predict(x_new, theta)

plt.scatter(x, y, color = "yellow")
plt.plot(x, y_predicted, color = "red")
plt.show()
```
