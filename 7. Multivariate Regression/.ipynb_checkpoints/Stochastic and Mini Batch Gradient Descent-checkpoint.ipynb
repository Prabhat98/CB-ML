```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np

x, y = make_regression(n_features = 1, n_samples = 5000, noise = 15, bias = 4) # what does the bias attribute do here ?
x = x.reshape((5000,1))
y = y.reshape((5000,1))

x = (x - x.mean())/(x.std())
plt.scatter(x, y, color = "orange")
plt.show()

print(x.shape,y.shape)
```

```python
def hypothesis(theta, x):
    y = theta[0] + theta[1]*x
    return y

def error(x, y, theta):
    total_error = 0
    m = x.shape[0]
    for i in range(m):
        total_error = total_error + ((y[i] - hypothesis(theta, x[i]))**2)
    return 0.5*total_error

def gradient(x, y, theta):
    grad = np.array([0.0,0.0])
    m = x.shape[0]
    for i in range(m):
        grad[0] = grad[0] + (hypothesis(theta, x[i]) - y[i])
        grad[1] = grad[1] + ((hypothesis(theta, x[i]) - y[i])*x[i])
    return grad

# batch_size = 1 means we are approximating gradient using one example means stochastic gradient descent, any batch_size>1 
# is mini batch gradient descent

# the error function does not comes out to be smooth, that is because the gradient we are calculating is not the 
# true gradient(the one obtained by iterating over all the data points and doing their sum). it is the gradient approximation,
# we are calculating the gradient of some data points and assuming it to be the gradient of the entire
# data set(which definetly will be different from the true gradient) and hence error will be generated over each update of 
# parameter. The error is being compensated as we are taking small steps(less number of iterations on batches,
# so less computation is required) in large number. Small steps(large in number) with less computations instead 
# of a big step(less in number) with larger computation is inexpensive 

# as the batch size increases the error curve becomes smoother and smoother(less noise/less oscillations/less vibrations/
# less zig-zag). we keep increasing batch size we will reach the batch gradient descent

def mini_batch_gradient_descent(x, y, theta, batch_size = 10):
    # instead of having consecutive batches we randomly shuffled all the elements of the array and then
    # picked first batch and passed to the loop
    m = x.shape[0]
    indices = np.arange(m)
    np.random.shuffle(indices)
    indice = indices[:batch_size]    
    grad = np.array([0.0,0.0])
    for i in indice:
        grad[0] = grad[0] + (hypothesis(theta, x[i]) - y[i])
        grad[1] = grad[1] + ((hypothesis(theta, x[i]) - y[i])*x[i])
    return grad

def gradient_descent_update(x, y, learning_rate, max_iterations):
    theta = np.array([0.0,0.0])
    combined_error = []
    for i in range(max_iterations):
        grad = mini_batch_gradient_descent(x, y, theta)
        err = error(x, y, theta)
        theta[0] = theta[0] - learning_rate*grad[0] 
        theta[1] = theta[1] - learning_rate*grad[1]
        combined_error.append(err)
    return theta, combined_error

theta, combined_error = gradient_descent_update(x, y, learning_rate = 0.1, max_iterations = 100)
print(theta)

y_predicted = hypothesis(theta, x)
plt.scatter(x, y, color = "orange")
plt.plot(x, y_predicted)
plt.show()

plt.plot(combined_error)
plt.show()
```
